##  Add categorical variables to Multiple Linear Regression 

### we can add categorical variables to Multiple Linear Regression using dummy variables and dropping one column for each category variable to make sure our matrices stays full rank to be able to calculate its inverse $$(X ′X) ^{−1} $$


```python
import numpy as np
import pandas as pd
import statsmodels.api as sm;
df = pd.read_csv('house_prices.csv')
display(df.head())

df_new=df.copy()
#create dummy variables for neighborhood
df_new[['A','B','C']]=pd.get_dummies(df.neighborhood)
#create dummy variables for style
df_new=df_new.join(  pd.get_dummies(df['style'])  )
df_new

#add intercept as statsmodels doesn't have it
df_new['intercept']=1
#train the model on neighborhood only using A as baseline
model=sm.OLS(df.price,df_new[['intercept','B','C']])
result=model.fit()
display(result.summary())

#plot the histogram of prices of the three neighborhood to check if the model coefficient really reflect the truth
plt.hist(df_new.query("C == 1")['price'], alpha = 0.3, label = 'C');
plt.hist(df_new.query("A == 1")['price'], alpha = 0.3, label = 'A');
plt.hist(df_new.query("B == 1")['price'], alpha = 0.3, label = 'B');
plt.legend();
#the coeff for B is 5.295e+05 which means it has a higher price than A  neighborhoods with this value which is clear
# in the histogram also c is less than A which is also clear in the histogram


#train the model on neighborhood, style,bathrooms and bedrooms using A  ans ranch as baselines
model=sm.OLS(df.price,df_new[['intercept','bedrooms','bathrooms','B','C','lodge','victorian']])
result=model.fit()
display(result.summary())
```

##  Using One_Zero_Negative_One Coding

### There is a key difference between the binary 1, 0 encodings that you've been working with, and the -1, 0, 1 encoding system: In the 1, 0 coding system, comparisons are to the baseline category as we've discussed, whereas in the 1, 0, -1 coding system comparisons are to the overall average.

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

df = pd.read_csv('house_prices.csv')
df2 = df.copy()
df.head()

## The below function creates 1, 0, -1 coded dummy variables.
def dummy_cat(df, col):
    '''
    INPUT:
    df - the dataframe where col is stored
    col - the categorical column you want to dummy (as a string)
    OUTPUT:
    df - the dataframe with the added columns
         for dummy variables using 1, 0, -1 coding
    '''
    for idx, val_0 in enumerate(df[col].unique()):
        if idx + 1 < df[col].nunique():            
            df[val_0] = df[col].apply(lambda x: 1 if x == val_0 else 0)
        else:    
            df[val_0] = df[col].apply(lambda x: -1 if x == val_0 else 0)
            for idx, val_1 in enumerate(df[col].unique()):
                if idx + 1 < df[col].nunique():
                    df[val_1] = df[val_0] + df[val_1]
                else:
                    del df[val_1]
    return df

new_df = dummy_cat(df, 'style') # Use on style
display(new_df.head(10))

new_df['intercept'] = 1

lm = sm.OLS(new_df['price'], new_df[['intercept', 'ranch', 'victorian']])
results = lm.fit()
results.summary()
```

#### Earlier, you fit linear models between each individual predictor variable and price, as well as using all of the variables and the price in a multiple linear regression model. Each of the individual models showed a positive relationship - that is, when bathrooms, bedrooms, or area increase, we predict the price of a home to increase. 

Fit a linear model to predict a home **price** using **bedrooms**, **bathrooms**, and **area**.

```python
#Fit a linear model to predict a home price using bedrooms, bathrooms, and area. 
df['intercept'] = 1
lm = sm.OLS(df['price'], df[['intercept', 'bedrooms', 'bathrooms','area']])
results = lm.fit()
results.summary()
```

### Calculate the VIFs for each variable in your model. [Here](https://etav.github.io/python/vif_factor_python.html) is the helpful post again, in case you need it!

```python
# get y and X dataframes based on this regression:
y, X = dmatrices('price ~  bedrooms + bathrooms + area'  , df, return_type='dataframe')
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["features"] = X.columns
vif.round(1)
```

### Interpreting Results of Logistic Regression

In this Cell we will be getting some practice with interpreting the coefficients in logistic regression. 

The dataset contains four variables: `admit`, `gre`, `gpa`, and `prestige`:

* `admit` is a binary variable. It indicates whether or not a candidate was admitted into UCLA (admit = 1) our not (admit = 0).
* `gre` is the GRE score. GRE stands for Graduate Record Examination.
* `gpa` stands for Grade Point Average.
* `prestige` is the prestige of an applicant alta mater (the school attended before applying), with 1 being the highest (highest prestige) and 4 as the lowest (not prestigious).



```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
#import data 
df = pd.read_csv("admissions.csv")
display(df.head())
pd.get_dummies(df.prestige)
#turn prestige into the dummy version.
df[['prest_1', 'prest_2', 'prest_3','prest_4',]] = pd.get_dummies(df['prestige'])
display(df)

#add intercept as statsmodels doesn't have it
df['intercept']=1
#train the model
model=sm.Logit(df['admit'],df[['intercept','gre','gpa','prest_2','prest_3','prest_4']])
result=model.fit()
display(result.summary2())



#now we exponentiaite parameters to get our interpretations
result_exp=np.exp(result.params)
print(result_exp)
print(1/result_exp)
```

## interpretation of Coefficients

## Logistic regression using sklearn


```python
import numpy as np 
import pandas as pd
from sklearn.linear_model import LogisticRegression as logR
from sklearn.model_selection import train_test_split as trn_tst
from sklearn.metrics import precision_score as precision , recall_score as recall,confusion_matrix as conf_matrix,accuracy_score as acc

df=pd.read_csv('admissions.csv')
display(df.head())
#CREATE DUMMY COLUMNS FOR PRESTIGE
df[['prest_1', 'prest_2', 'prest_3', 'prest_4']] = pd.get_dummies(df['prestige'])
#DEFINE INPUT, OUTPUT
x = df.drop(['admit', 'prestige', 'prest_1'] , axis=1)
y = df['admit']
#SPLIT TRAIN AND TEST
x_tr,x_tst,y_tr,y_tst=trn_tst(x,y,test_size=0.2,random_state=0)
#TRAIN MODEL
model=logR()
model.fit(x_tr,y_tr)
#TRY MODEL
y_pre=model.predict(x_tst)
#GET METRICS
print('accuracy: ',acc(y_tst,y_pre))
print('accuracy: ',acc(y_tst,y_pre))
print('Recall: ',recall(y_tst,y_pre))
print('precision: ',precision(y_tst,y_pre))
print('conf_matrix: ',pd.DataFrame(conf_matrix(y_tst,y_pre)))
```
